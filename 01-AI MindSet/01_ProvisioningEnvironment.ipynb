## 🏗️ Step 1: Core Runtimes

We need two engines to run our ecosystem:
1. **Node.js (v18+):** Powers the Claude Code CLI and the Model Context Protocol (MCP).
2. **Python (3.10+):** Powers our Multi-Agent Swarms and logic processing.

Run the cell below to verify your engines.

````python
import subprocess

def verify_runtime(command, name):
    try:
        version = subprocess.check_output([command, "--version"], text=True).strip()
        print(f"✅ {name} detected: {version}")
    except Exception:
        print(f"❌ {name} NOT FOUND. Please install it before proceeding.")

verify_runtime("node", "Node.js")
verify_runtime("python3", "Python 3")
````

## 🔐 Step 3: Secret Management & Guardrails

Vibe Engineering requires strict security. We must ensure the AI never leaks our secrets.

**1. Create a `.env` file for your keys.**
**2. Create a `.gitignore` to protect that file.**

Run the cell below to generate a safe boilerplate for this project.

````python
import os

gitignore_content = """
.env
__pycache__/
.ipynb_checkpoints/
node_modules/
dist/
"""

with open(".gitignore", "w") as f:
    f.write(gitignore_content.strip())

if not os.path.exists(".env"):
    with open(".env", "w") as f:
        f.write("ANTHROPIC_API_KEY=your_key_here\n")
    print("✅ Created .gitignore and .env (template).")
else:
    print("✅ .env already exists. Skipping creation to protect your keys.")
````
